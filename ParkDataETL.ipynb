```python
import pandas as pd
park = pd.read_csv('C:\Roman\code\Python ETL\park-data.csv', encoding= 'unicode_escape')
park = park.drop(['SquirrelSighter(s)','OtherAnimal Sightings'],axis=1)
if park['ParkID'].dtypes == float:
    park = park.astype({'ParkID':int})
    print("Data type changed")



if len(park['ParkID'].unique()) < len(park.index):
    print('Duplicate value is present Remove it..')
    park.sort_values('ParkID',inplace=True)
    park.drop_duplicates(subset='ParkID',keep='first',inplace=True)
    print("Duplicate second value removed")


if pd.isnull(park['TemperatureWeather']).any():
    park['TemperatureWeather'].fillna('43 degrees, sunny',inplace = True)
    park[['Temperature', 'Weather']] = park.TemperatureWeather.str.split(pat = ',',n=1,expand=True)
#park[['Temperatur', 'Weather']] = park["TemperatureWeather"].apply(lambda x: pd.Series(str(x).split(",")))

park = park.drop('TemperatureWeather',axis=1)
park.head()





```
