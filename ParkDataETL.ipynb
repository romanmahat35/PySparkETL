```python
import pandas as pd
park = pd.read_csv('C:\Roman\code\Python ETL\park-data.csv', encoding= 'unicode_escape')
park = park.drop(['SquirrelSighter(s)','OtherAnimal Sightings'],axis=1)
print('Before Processing a Actual File')
park.head(2)
```

Here I have Perfrom Basic Tranformation To Park Data 
    1.Drop the unused Columns
    2.Change the data type of columns
    3.Check for Duplicate of data using primary key if dupilcate data is there Deleting the 
      first row
    4.Replace a Null value with Most frequetly comming value
    5.split a temperture and weather column from temperturewether column and  drop temperturewether column

```python
if park['ParkID'].dtypes == float:
    park = park.astype({'ParkID':int})
    print("Data type changed")



if len(park['ParkID'].unique()) < len(park.index):
    print('Duplicate value is present Remove it..')
    park.sort_values('ParkID',inplace=True)
    park.drop_duplicates(subset='ParkID',keep='first',inplace=True)
    print("Duplicate second value removed")


if pd.isnull(park['TemperatureWeather']).any():
    park['TemperatureWeather'].fillna('43 degrees, sunny',inplace = True)
    park[['Temperature', 'Weather']] = park.TemperatureWeather.str.split(pat = ',',n=1,expand=True)
#park[['Temperatur', 'Weather']] = park["TemperatureWeather"].apply(lambda x: pd.Series(str(x).split(",")))
park['Litter'].fillna('Some',inplace=True)
park.head(3)
 
```

```python
#concat date and starttime and date and endtime 
park['StartDatetime']  = park['Date'] + ' ' + park['StartTime'] 
park['EndDateTime'] = park['Date'] + ' ' + park['EndTime']
park = park.drop(['Date','StartTime','EndTime','TemperatureWeather'],axis=1)
park['AreaName'] = park['AreaName'].str.lower()
park.head(2)
```
